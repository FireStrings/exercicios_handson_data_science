```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import sklearn
import seaborn as sns
%matplotlib inline
from sklearn.model_selection import StratifiedShuffleSplit
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
```

Dados de PIB / Happiness

```python
df_bli = pd.read_csv("data/BLI_25032023163010840.csv")
```

### Transform data
Here we need to do some transformations, like the "Total" selection only, the transpose and some rename.

```python
def transform_data(df):
    l_df = {}
    for i in enumerate(df.Country.unique()):
        index, name = i
        df_country = df_bli\
        .query("Country == '"+name+"' & Inequality == 'Total'")[["Indicator", "Unit", "Value"]]\
        .head(200).T
        
        df_country = df_country.T.rename(columns={"Value": name})
        l_df[index] = df_country
    
    for i in l_df:
        if i == 0:
            df_base = l_df[i]
            continue
            
        df_x = df_base
        df_y = l_df[i]
        
        df_base = df_x.merge(df_y, left_on='Indicator', right_on='Indicator', how="outer")
        count_null_x = df_base[df_base["Unit_x"].isnull()].Indicator.count()
        count_null_y = df_base[df_base["Unit_y"].isnull()].Indicator.count()
        
        if count_null_x > 0:
            df_base = df_base.drop(["Unit_x"], axis=1).fillna(0).rename(columns={"Unit_y": "Unit"})
        elif count_null_y > 0:
            df_base = df_base.drop(["Unit_y"], axis=1).fillna(0).rename(columns={"Unit_x": "Unit"})
        else:
            df_base = df_base.drop(["Unit_y"], axis=1).fillna(0).rename(columns={"Unit_x": "Unit"})
    
    
    return df_base[["Indicator", "Unit", *df.Country.unique()]]
    
df = transform_data(df_bli)
```

```python
df_transpose = df.set_index("Indicator").drop(["Unit"], axis=1).T
```

### Countries with the highest life satisfaction

```python
df_transpose.sort_values(by=['Life satisfaction'], ascending=False).head(10)
```

### Normalization / Scaling
Here we need to scaling the data

```python
df_norm = (df_transpose-df_transpose.min())/(df_transpose.max()-df_transpose.min())
df_norm.head()
```
